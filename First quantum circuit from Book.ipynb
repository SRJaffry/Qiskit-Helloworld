In this code, first I create a single-qubit quantum circuit. I initialize its state to be 1/sqrt(2) each for state 0 and 1. <br>
Then I execute it via simulator and observe the histogram to see that each state exist for 0.5 probability. <br>

Then I do the measurement. Measurement actually collapse the qubit in one of the state with certain probability of 1 <br>

This is the code that I used from the book.

```python
from qiskit import *
from qiskit import execute, Aer
from qiskit.tools.visualization import plot_histogram
from math import *
```

```python
# First make a Quantum circuit with one qubit qc = QuantumCircuit(#Qubits)
qc = QuantumCircuit(1)

# Set initial states to be 1/sqrt(2) for each state. 
# Qubit exist in either state with probability 1/sqrt(2). 
# This is called superposition state. 
initial_state = [1/sqrt(2),1/sqrt(2)]
qc.initialize(initial_state)
qc.draw(output = 'mpl', initial_state=True)
```

```python
# Call the Quantum simulator on local machine. 
backend = Aer.get_backend('statevector_simulator')

# Execute the circuit and see the results.
result = execute(qc,backend).result()
counts = result.get_counts()
plot_histogram(counts)
```

### Perform measurement

All qubits are in probabilistic states generally. 
It is only when they are measured that we can say that qubits are in certain state with Probability 1 and not in the other state with probability 1. 

Hence, in the following, we will do the measurement and observe teh states and their probabilities

```python
qc.measure_all()
result = execute(qc,backend).result()
counts = result.get_counts()

plot_histogram(counts)
```
