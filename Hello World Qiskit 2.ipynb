Click [here](https://qiskit.org/textbook/ch-states/atoms-computation.html)

```python
# We start by creating a circuit with eight qubits and eight outputs. 
# The QuantumCircuit takes the number of qubits in the quantum circuit as an argument.
qc_output = QuantumCircuit(8)

# The extraction of outputs in a quantum circuit is done using an operation called measure_all().
qc_output.measure_all()
```

```python
qc_output.draw(output = 'mpl', initial_state = True) 
```

```python
sim = Aer.get_backend('aer_simulator') 
```

```python
result = sim.run(qc_output).result()
counts = result.get_counts()
```

```python
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
